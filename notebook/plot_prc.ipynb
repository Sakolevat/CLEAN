```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

#### Uniref1(10) results

```python
pencil_precision = np.zeros((5,100))
pencil_recall = np.zeros((5,100))

for i in range(5):
    result_file = open('../eval/uniref1/triplet/uniref1_test_split_' + str(i) + '_curate_prc_result.csv','r')
    csvreader = csv.reader(result_file, delimiter = ',')
    for j, rows in enumerate(csvreader):
        if j > 0:
            pencil_precision[i][j-1] = float(rows[1])
            pencil_recall[i][j-1] = float(rows[2])
##################################################################################
esm_precision = np.zeros((5,100))
esm_recall = np.zeros((5,100))

for i in range(5):
    result_file = open('../eval/uniref1/esm/uniref1_test_split_' + str(i) + '_curate_prc_esm_result.csv','r')
    csvreader = csv.reader(result_file, delimiter = ',')
    for j, rows in enumerate(csvreader):
        if j > 0:
            esm_precision[i][j-1] = float(rows[1])
            esm_recall[i][j-1] = float(rows[2])
##################################################################################
supcon_precision = np.zeros((5,100))
supcon_recall = np.zeros((5,100))

for i in range(5):
    result_file = open('../eval/uniref1/supcon/uniref1_test_split_' + str(i) + '_curate_prc_result.csv','r')
    csvreader = csv.reader(result_file, delimiter = ',')
    for j, rows in enumerate(csvreader):
        if j > 0:
            supcon_precision[i][j-1] = float(rows[1])
            supcon_recall[i][j-1] = float(rows[2])


##################################################################################
plt.plot(np.mean(supcon_recall, axis=0), np.mean(supcon_precision, axis=0), label = 'PENCiL-supcon')
plt.plot(np.mean(pencil_recall, axis=0), np.mean(pencil_precision, axis=0), label = 'PENCiL-triplet')
plt.plot(np.mean(esm_recall, axis=0), np.mean(esm_precision, axis=0), label = 'ESM-1b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision recall curve Uniref10')
plt.legend()
plt.show()
```
