```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

#### Uniref10 Prom results

```python
pencil_precision = np.zeros((1,100))
pencil_recall = np.zeros((1,100))

result_file = open('../eval/uniref10_prom_split_test_curate_prc_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        pencil_precision[0][j-1] = float(rows[1])
        pencil_recall[0][j-1] = float(rows[2])
```

```python
esm_precision = np.zeros((1,100))
esm_recall = np.zeros((1,100))

result_file = open('../eval/uniref10_prom_split_test_curate_prc_esm_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        esm_precision[0][j-1] = float(rows[1])
        esm_recall[0][j-1] = float(rows[2])
```

```python
plt.plot(np.mean(pencil_recall, axis=0), np.mean(pencil_precision, axis=0), label = 'PENCiL')
plt.plot(np.mean(esm_recall, axis=0), np.mean(esm_precision, axis=0), label = 'ESM-1b')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision recall curve Uniref50 Promiscuous Enzymes')
plt.legend()
plt.show()
```
