```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

#### Uniref1(10) subset results [2,5,10,25,50]

```python
subset_pencil_precision = []
subset_pencil_recall = []

for bin in range(5):
    pencil_precision = np.zeros((5,100))
    pencil_recall = np.zeros((5,100))
    for i in range(5):
        #uniref10_test_split_0_curate_subset0_prc_result
        result_file = open(
            '../eval/subset/uniref1/triplet/uniref1_test_split_' 
            + str(i) + '_curate_subset' + str(bin) +'_prc_result.csv','r')
        csvreader = csv.reader(result_file, delimiter = ',')
        for j, rows in enumerate(csvreader):
            if j > 0:
                pencil_precision[i][j-1] = float(rows[1])
                pencil_recall[i][j-1] = float(rows[2])
    subset_pencil_precision.append(pencil_precision)
    subset_pencil_recall.append(pencil_recall)
#################################################################################
subset_esm_precision = []
subset_esm_recall = []

for bin in range(5):
    esm_precision = np.zeros((5,100))
    esm_recall = np.zeros((5,100))
    for i in range(5):
        #uniref10_test_split_0_curate_subset0_prc_result
        result_file = open(
            '../eval/subset/uniref1/esm/uniref1_test_split_' 
            + str(i) + '_curate_subset' + str(bin) +'_prc_esm_result.csv','r')
        csvreader = csv.reader(result_file, delimiter = ',')
        for j, rows in enumerate(csvreader):
            if j > 0:
                esm_precision[i][j-1] = float(rows[1])
                esm_recall[i][j-1] = float(rows[2])
    subset_esm_precision.append(esm_precision)
    subset_esm_recall.append(esm_recall)
#################################################################################
subset_supcon_precision = []
subset_supcon_recall = []

for bin in range(5):
    supcon_precision = np.zeros((3,100))
    supcon_recall = np.zeros((3,100))
    for i in range(3):
        #uniref10_test_split_0_curate_subset0_prc_result
        result_file = open(
            '../eval/subset/uniref1/supcon/uniref1_test_split_' 
            + str(i) + '_curate_subset' + str(bin) +'_prc_result.csv','r')
        csvreader = csv.reader(result_file, delimiter = ',')
        for j, rows in enumerate(csvreader):
            if j > 0:
                supcon_precision[i][j-1] = float(rows[1])
                supcon_recall[i][j-1] = float(rows[2])
    subset_supcon_precision.append(supcon_precision)
    subset_supcon_recall.append(supcon_recall)
```

```python
fig = plt.figure(figsize=(16, 8))

plt.plot(np.mean(subset_pencil_recall[0], axis=0), np.mean(subset_pencil_precision[0], axis=0), label = 'PENCiL [1,5) ')
#plt.plot(np.mean(subset_esm_recall[0], axis=0), np.mean(subset_esm_precision[0], axis=0), '--', label = 'ESM-1b [1,5)')
plt.plot(np.mean(subset_supcon_recall[0], axis=0), np.mean(subset_supcon_precision[0], axis=0), '-.', label = 'supcon [1,5)')

plt.plot(np.mean(subset_pencil_recall[1], axis=0), np.mean(subset_pencil_precision[1], axis=0), label = 'PENCiL [5,10)')
#plt.plot(np.mean(subset_esm_recall[1], axis=0), np.mean(subset_esm_precision[1], axis=0), '--', label = 'ESM-1b [5,10)')
plt.plot(np.mean(subset_supcon_recall[1], axis=0), np.mean(subset_supcon_precision[1], axis=0), '-.', label = 'supcon [5,10)')

plt.plot(np.mean(subset_pencil_recall[2], axis=0), np.mean(subset_pencil_precision[2], axis=0), label = 'PENCiL [10,25)')
#plt.plot(np.mean(subset_esm_recall[2], axis=0), np.mean(subset_esm_precision[2], axis=0), '--', label = 'ESM-1b [10,25)')
plt.plot(np.mean(subset_supcon_recall[2], axis=0), np.mean(subset_supcon_precision[2], axis=0), '-.', label = 'supcon [10,25)')

plt.plot(np.mean(subset_pencil_recall[3], axis=0), np.mean(subset_pencil_precision[3], axis=0), label = 'PENCiL [25,50)')
#plt.plot(np.mean(subset_esm_recall[3], axis=0), np.mean(subset_esm_precision[3], axis=0), '--', label = 'ESM-1b [25,50)')
plt.plot(np.mean(subset_supcon_recall[3], axis=0), np.mean(subset_supcon_precision[3], axis=0), '-.', label = 'supcon [25,50)')

plt.plot(np.mean(subset_pencil_recall[4], axis=0), np.mean(subset_pencil_precision[4], axis=0), label = 'PENCiL [50,$\infty$)')
#plt.plot(np.mean(subset_esm_recall[4], axis=0), np.mean(subset_esm_precision[4], axis=0), '--', label = 'ESM-1b [50,$\infty$)')
plt.plot(np.mean(subset_supcon_recall[4], axis=0), np.mean(subset_supcon_precision[4], axis=0), '-.', label = 'supcon [50,$\infty$)')


plt.xlabel('Recall')
plt.ylabel('Precision')

plt.title('Precision recall curve Uniref50, subsetted on EC frequency')
plt.legend()
plt.show()
```
