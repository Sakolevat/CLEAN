```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

### All Prom results (-p 0.0001 -up 0.05 -s 100)

```python
pencil_precision100 = np.zeros((1,200))
pencil_recall100 = np.zeros((1,200))

result_file = open('../eval/prom100/uniref100_prom_split_test_curate_prc_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        pencil_precision100[0][j-1] = float(rows[1])
        pencil_recall100[0][j-1] = float(rows[2])
###############################################################################################        
esm_precision100 = np.zeros((1,200))
esm_recall100 = np.zeros((1,200))

result_file = open('../eval/prom100/uniref100_prom_split_test_curate_prc_esm_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        esm_precision100[0][j-1] = float(rows[1])
        esm_recall100[0][j-1] = float(rows[2])
###############################################################################################  
pencil_precision10 = np.zeros((1,200))
pencil_recall10 = np.zeros((1,200))

result_file = open('../eval/prom10/uniref10_prom_split_test_curate_prc_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        pencil_precision10[0][j-1] = float(rows[1])
        pencil_recall10[0][j-1] = float(rows[2])        
###############################################################################################
esm_precision10 = np.zeros((1,200))
esm_recall10 = np.zeros((1,200))

result_file = open('../eval/prom10/uniref10_prom_split_test_curate_prc_esm_result.csv','r')
csvreader = csv.reader(result_file, delimiter = ',')
for j, rows in enumerate(csvreader):
    if j > 0:
        esm_precision10[0][j-1] = float(rows[1])
        esm_recall10[0][j-1] = float(rows[2])
###############################################################################################  
plt.plot(np.mean(pencil_recall100, axis=0), np.mean(pencil_precision100, axis=0), label = 'PENCiL-Uniref100')
plt.plot(np.mean(pencil_recall10, axis=0), np.mean(pencil_precision10, axis=0), label = 'PENCiL-Uniref50')
plt.plot(np.mean(esm_recall100, axis=0), np.mean(esm_precision100, axis=0), label = 'ESM-Uniref100')
plt.plot(np.mean(esm_recall10, axis=0), np.mean(esm_precision10, axis=0), label = 'ESM-Uniref50')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision recall curve for Promiscuous Enzymes, p-value from 0.0001 to 0.01')
plt.legend()
plt.show()
```
