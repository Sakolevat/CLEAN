```python
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
subset_pencil_precision = []
subset_pencil_recall = []

for bin in range(5):
    pencil_precision = np.zeros((5,100))
    pencil_recall = np.zeros((5,100))
    for i in range(5):
        #uniref10_test_split_0_curate_subset0_prc_result
        result_file = open('../eval/subset/subset1to5/uniref10_test_split_' + str(i) + '_curate_subset' + str(bin) +'_prc_result.csv','r')
        csvreader = csv.reader(result_file, delimiter = ',')
        for j, rows in enumerate(csvreader):
            if j > 0:
                pencil_precision[i][j-1] = float(rows[1])
                pencil_recall[i][j-1] = float(rows[2])
    subset_pencil_precision.append(pencil_precision)
    subset_pencil_recall.append(pencil_recall)
```

```python
subset_esm_precision = []
subset_esm_recall = []

for bin in range(5):
    esm_precision = np.zeros((5,100))
    esm_recall = np.zeros((5,100))
    for i in range(5):
        #uniref10_test_split_0_curate_subset0_prc_result
        result_file = open('../eval/subset/subset1to5/uniref10_test_split_' + str(i) + '_curate_subset' + str(bin) +'_prc_esm_result.csv','r')
        csvreader = csv.reader(result_file, delimiter = ',')
        for j, rows in enumerate(csvreader):
            if j > 0:
                esm_precision[i][j-1] = float(rows[1])
                esm_recall[i][j-1] = float(rows[2])
    subset_esm_precision.append(esm_precision)
    subset_esm_recall.append(esm_recall)
```

```python
fig = plt.figure(figsize=(16, 8))

plt.plot(np.mean(subset_pencil_recall[0], axis=0), np.mean(subset_pencil_precision[0], axis=0), label = 'PENCiL [2,5) ')
plt.plot(np.mean(subset_esm_recall[0], axis=0), np.mean(subset_esm_precision[0], axis=0), '--', label = 'ESM-1b [2,5)')

plt.plot(np.mean(subset_pencil_recall[1], axis=0), np.mean(subset_pencil_precision[1], axis=0), label = 'PENCiL [5,20)')
plt.plot(np.mean(subset_esm_recall[1], axis=0), np.mean(subset_esm_precision[1], axis=0), '--', label = 'ESM-1b [5,20)')

plt.plot(np.mean(subset_pencil_recall[2], axis=0), np.mean(subset_pencil_precision[2], axis=0), label = 'PENCiL [20,50)')
plt.plot(np.mean(subset_esm_recall[2], axis=0), np.mean(subset_esm_precision[2], axis=0), '--', label = 'ESM-1b [20,50)')

plt.plot(np.mean(subset_pencil_recall[3], axis=0), np.mean(subset_pencil_precision[2], axis=0), label = 'PENCiL [50,100)')
plt.plot(np.mean(subset_esm_recall[3], axis=0), np.mean(subset_esm_precision[2], axis=0), '--', label = 'ESM-1b [50,100)')

plt.plot(np.mean(subset_pencil_recall[4], axis=0), np.mean(subset_pencil_precision[2], axis=0), label = 'PENCiL [100,$\infty$)')
plt.plot(np.mean(subset_esm_recall[4], axis=0), np.mean(subset_esm_precision[2], axis=0), '--', label = 'ESM-1b [100,$\infty$)')

plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision recall curve Uniref10, subsetted on EC frequency')
plt.legend()
plt.show()
```
